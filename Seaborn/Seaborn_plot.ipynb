```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
# Scatter Plot
plt.figure(figsize=(6,4))
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot of Total Bill by Day')
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(6,4))
sns.violinplot(x='day', y='total_bill', data=tips, hue= "sex", split=True)
plt.title('Violin Plot of Total Bill by Day')
plt.show()
```

```python
# Pair Plot
tips_df = sns.load_dataset('tips')
sns.pairplot(tips_df, hue='sex', palette='Blues')
```

```python
# Heatmap of Correlation Matrix
iris_numeric = iris.select_dtypes(include=[float, int])

# Calculate correlation matrix
corr = iris_numeric.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='Blues', linecolor='b', linewidth=1)
plt.title('Heatmap of Iris Correlation Matrix')
plt.show()
```

```python
# Explain the difference between a box plot and a violin plot. In what scenarios would you choose one over the other?
# Use the tips dataset to illustrate your explanation with code.
```

Box Plot: A box plot visualizes the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It also shows outliers. It is useful for detecting skewness and spotting outliers.

Violin Plot: A violin plot is a combination of a box plot and a kernel density estimate (KDE). It provides the same summary statistics as a box plot but also shows the distribution shape of the data (smoothed probability density).

Scenario: If you want to quickly compare summary statistics across different categories, use a box plot. If you are interested in understanding the distribution's shape in more detail (e.g., bimodality), use a violin plot.

```python
tips = sns.load_dataset('tips')

fig = plt.figure(figsize=(12, 5))

# Box Plot
ax1 = fig.add_subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
sns.boxplot(x='day', y='total_bill', data=tips, ax=ax1)
ax1.set_title('Box Plot of Total Bill by Day')

# Violin Plot
ax2 = fig.add_subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
sns.violinplot(x='day', y='total_bill', data=tips, ax=ax2)
ax2.set_title('Violin Plot of Total Bill by Day')

plt.tight_layout()
plt.show()
```

```python
# What is the purpose of using regplot in Seaborn? How does it differ from using scatterplot?
# Provide an example using the tips dataset to show how regression lines can provide insights into data trends.
```

regplot: Adds a regression line (along with a confidence interval) to the scatter plot. It is used when you want to visualize both the data points and the linear trend in the data.

scatterplot: Only shows the data points without any fitted line.

Difference: Use scatterplot for purely visualizing relationships, while regplot is preferred when you want to see the underlying trend (with a regression model).

```python
fig = plt.figure(figsize=(12, 5))

# Scatter Plot
ax1 = fig.add_subplot(1, 2, 1)  
sns.scatterplot(x='total_bill', y='tip', data=tips, ax=ax1)
ax1.set_title('Scatter Plot: Total Bill vs Tip')

# Regression Plot
ax2 = fig.add_subplot(1, 2, 2) 
sns.regplot(x='total_bill', y='tip', data=tips, ax=ax2)
ax2.set_title('Regression Plot: Total Bill vs Tip')

plt.tight_layout()
plt.show()
```

For visualizing missing data effects, Seaborn’s heatmap can highlight where missing values occur.

```python
titanic = sns.load_dataset('titanic')

# Heatmap to visualize missing data
plt.figure(figsize=(8,6))
sns.heatmap(titanic.isnull(), cbar=False, cmap='inferno')
plt.title('Missing Data in Titanic Dataset')
plt.show()
```
