```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

import pandas as pd
import numpy as np
import seaborn as sns
import chart_studio.plotly as py
import cufflinks as cf
import plotly.express as px

%matplotlib inline

# Make Plotly work in your Jupyter Notebook
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
# Use Plotly locally
cf.go_offline()
```

```python
df = px.data.iris()  
```

```python
# Scatter Plot

scatter_fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species",
                         title="Scatter Plot: Sepal Width vs Length")
scatter_fig.update_layout(width=800, height=600)
scatter_fig.show()
```

```python
# Line Plot

line_fig = px.line(df, x="sepal_width", y="sepal_length", color="species",
                   title="Line Plot: Sepal Length vs Width")
line_fig.update_layout(width=800, height=600)
line_fig.show()
```

```python
# Bar Plot

bar_fig = px.bar(df, x="species", y="sepal_width", title="Bar Plot: Sepal Width by Species",
                 color="species", barmode="group")
bar_fig.update_layout(width=800, height=600)
bar_fig.show()
```

```python
# Histogram

histogram_fig = px.histogram(df, x="sepal_width", nbins=20, title="Histogram: Sepal Width Distribution")
histogram_fig.update_layout(width=800, height=600)
histogram_fig.show()
```

```python
# Box Plot

box_fig = px.box(df, x="species", y="sepal_width", title="Box Plot: Sepal Width by Species")
box_fig.update_layout(width=800, height=600)
box_fig.show()
```

```python
# Violin Plot

violin_fig = px.violin(df, x="species", y="sepal_width", box=True, points="all",
                       title="Violin Plot: Sepal Width by Species")
violin_fig.update_layout(width=800, height=600)
violin_fig.show()
```

```python
# Pie Chart

pie_fig = px.pie(df, names='species', title="Pie Chart: Species Distribution")
pie_fig.update_layout(width=800, height=600)
pie_fig.show()
```

```python
# Density Heatmap

heatmap_fig = px.density_heatmap(df, x="sepal_width", y="sepal_length", 
                                 title="Density Heatmap: Sepal Width vs Length")
heatmap_fig.update_layout(width=800, height=600)
heatmap_fig.show()
```

```python
# Customizing Axes, Titles, and Fonts

custom_fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species")
custom_fig.update_layout(
    title="Customized Scatter Plot",
    xaxis_title="Sepal Width (cm)",
    yaxis_title="Sepal Length (cm)",
    legend_title="Species",
    font=dict(family="Arial", size=18, color="DarkBlue")
)
custom_fig.update_layout(width=800, height=600)
custom_fig.show()
```

```python
# Adding Annotations

annotated_fig = go.Figure(data=go.Scatter(x=df['sepal_width'], y=df['sepal_length'], 
                                          mode='markers', marker=dict(color=df['species_id'])))
annotated_fig.update_layout(title="Scatter Plot with Annotations")
annotated_fig.add_annotation(x=3, y=5, text="Sample Annotation", showarrow=True)
annotated_fig.update_layout(width=800, height=600)
annotated_fig.show()
```

```python
# Interactive Hover Info

hover_fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species",
                       hover_data=['petal_width', 'petal_length'],
                       title="Scatter Plot with Hover Info")
hover_fig.update_layout(width=800, height=600)
hover_fig.show()
```

```python
# Adding Buttons for Plot Interactions (Dropdowns)

button_fig = go.Figure()

button_fig.add_trace(go.Scatter(x=df['sepal_width'], y=df['sepal_length'], mode='markers', 
                                marker=dict(color='blue'), name='All Points'))

# Adding buttons for filtering
button_fig.update_layout(
    updatemenus=[
        dict(
            buttons=list([
                dict(label="Show All", method="update", args=[{"visible": [True, False]}]),
                dict(label="Show Points", method="update", args=[{"visible": [False, True]}])
            ]),
            direction="down", showactive=True
        )
    ]
)
button_fig.update_layout(width=800, height=600)
button_fig.show()
```

```python
# Subplots and Faceting

from plotly.subplots import make_subplots

subplot_fig = make_subplots(rows=1, cols=2)

subplot_fig.add_trace(go.Scatter(x=df['sepal_width'], y=df['sepal_length'], 
                                 mode='markers', name='Sepal'), row=1, col=1)

subplot_fig.add_trace(go.Scatter(x=df['petal_width'], y=df['petal_length'], 
                                 mode='markers', name='Petal'), row=1, col=2)

subplot_fig.update_layout(title="Subplots: Sepal and Petal")
subplot_fig.update_layout(width=800, height=600)
subplot_fig.show()
```

```python
# Faceting in Plotly Express

facet_fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species", 
                       facet_col="species", title="Faceted Scatter Plot by Species")
facet_fig.update_layout(width=800, height=600)
facet_fig.show()
```

```python
# Heatmaps and Correlation Matrices

correlation_matrix = df.iloc[:,0:4].corr()

heatmap_fig = go.Figure(data=go.Heatmap(
    z=correlation_matrix.values, 
    x=correlation_matrix.columns, 
    y=correlation_matrix.columns, 
    colorscale='Viridis'
))

heatmap_fig.update_layout(title="Heatmap: Feature Correlation Matrix")
heatmap_fig.update_layout(width=800, height=600)
heatmap_fig.show()
```

```python
# Animations

animated_fig = px.scatter(df, x="sepal_width", y="sepal_length", 
                          color="species", animation_frame="species", 
                          title="Animated Scatter Plot")
animated_fig.update_layout(width=800, height=600)
animated_fig.show()
```

```python
# 3D Plots

scatter_3d_fig = px.scatter_3d(df, x="sepal_width", y="sepal_length", z="petal_width", 
                               color="species", title="3D Scatter Plot")
scatter_3d_fig.update_layout(width=800, height=600)
scatter_3d_fig.show()
```

```python
# 3D Line Plot

line_3d_fig = px.line_3d(df, x="sepal_width", y="sepal_length", z="petal_length", 
                         color="species", title="3D Line Plot")
line_3d_fig.update_layout(width=800, height=600)
line_3d_fig.show()
```
