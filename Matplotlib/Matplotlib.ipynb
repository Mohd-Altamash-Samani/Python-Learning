```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# 1. Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

# plt.subplot(3, 2, 2)  # Row 3, Column 2, Position 2
plt.figure(figsize=(4, 3))

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
```

```python
# 2. Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 100
colors = np.random.rand(50)

plt.figure(figsize=(4, 3))
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5, cmap='viridis')
plt.colorbar()  # Show color scale
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
```

```python
# 3. Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 40]

plt.figure(figsize=(4, 3))
plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
```

```python
# 4. Histogram
data = np.random.randn(1000)

plt.figure(figsize=(4, 3))
plt.hist(data, bins=30, edgecolor='black', color='orange')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
```

```python
# 5. Subplots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(4, 3))
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Subplots')
```

```python
# 6. Box Plot
data = [np.random.rand(50), np.random.rand(50), np.random.rand(50)]

plt.figure(figsize=(4, 3))
plt.boxplot(data, notch=True, patch_artist=True)
plt.xticks([1, 2, 3], ['Group 1', 'Group 2', 'Group 3'])
plt.ylabel('Value')
plt.title('Box Plot')
```

```python
# 7. Stacked Bar Plot
category_names = ['A', 'B', 'C', 'D']
group_data = np.array([[10, 15, 20, 25], [15, 20, 25, 30]])

plt.figure(figsize=(4, 3))
plt.bar(category_names, group_data[0], label='Group 1', color='skyblue')
plt.bar(category_names, group_data[1], bottom=group_data[0], label='Group 2', color='orange')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Plot')
plt.legend()

# Adjust layout for better spacing
plt.tight_layout()

# Show the figure containing all plots
plt.show()
```

```python
# 8. Violin Plot
plt.figure(figsize=(4, 3))
sns.violinplot(data=np.random.randn(100, 3), palette="Set3", inner="point")
plt.title('Violin Plot')
plt.show()
```

```python
# 9. Heatmap
data = np.random.rand(10, 12)
plt.figure(figsize=(5, 4))
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title('Heatmap Example')
plt.show()
```

```python
# 10. Pie Chart
labels = ['Group A', 'Group B', 'Group C']
sizes = [50, 35, 15]
colors = ['gold', 'lightcoral', 'lightskyblue']

plt.figure(figsize=(4, 3))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart Example')
plt.show()
```

```python
# 11. Annotating a Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(4, 3))
plt.plot(x, y)
plt.annotate('Max point', xy=(np.pi/2, 1), xytext=(np.pi, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot with Annotation')
plt.show()
```

```python
# 12. Error Bars
x = np.linspace(0, 10, 50)
y = np.sin(x)
yerr = 0.1

plt.figure(figsize=(4, 3))
plt.errorbar(x, y, yerr=yerr, fmt='-o', capsize=5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Error Bars Example')
plt.show()
```

```python
# 13. Multiple Lines in One Plot
x = np.linspace(0, 2, 100)
plt.figure(figsize=(4, 3))
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title("Multiple Lines")
plt.legend()
plt.show()
```

```python
# 14. Saving Plot Example
plt.figure(figsize=(4, 3))
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title("Multiple Lines")
plt.legend()
# plt.show()
plt.savefig('save_plot.png', dpi=300, bbox_inches='tight')
# plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Basic Graph
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(6,4), dpi=100)

# Line 1
# Keyword Argument Notation
#plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')
# Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--', label='2x')

## Line 2
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')

# Plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')
plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([0,1,2,3,4,])
plt.yticks([0,2,4,6,8,10,12,14,16])
plt.legend()
plt.show()
```

```python
# Bar Chart
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(6,4), dpi=100)
bars = plt.bar(labels, values)
patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.show()
```

```python
# Histogram
fifa = pd.read_csv('fifa_data.csv')

plt.figure(figsize=(6,4), dpi=100)
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins=bins, color='#abcdef')

plt.xticks(bins)
plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')
plt.show()
```

```python
# Pie Chart 1
plt.figure(figsize=(6,4), dpi=100)

plt.style.use("classic")

left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']
plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%')
plt.title('Foot Preference of FIFA Players')
plt.show()
```

```python
# Pie Chart 2
plt.figure(figsize=(6,4), dpi=100)
plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.0,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')
plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```

```python
# Box and Whiskers Chart
plt.figure(figsize=(6,4), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    box.set(hatch = '/')
    
plt.show()
```

```python
import pandas as pd
df=pd.read_csv("gas_prices.csv")
df.head(1)
```

```python
# Line Plot for a Single Country:
# Create a line plot to visualize the inflation rate trend for Australia over the years. 
# How do you label the x-axis, y-axis, and add a title to the plot?

plt.figure(figsize=(6,4),dpi=100)
plt.plot(df['Year'], df['Australia'], marker='o')

plt.title('Inflation Rate in Australia (1990-1994)')
plt.xlabel('Year')
plt.ylabel('Inflation Rate (%)')

plt.grid(True)
plt.show()
```

```python
# Multiple Line Plot for Comparison:
# Plot the inflation rates for Australia, Canada, and France on the same graph to compare their trends from 1990 to 1994. 
# How would you distinguish each country’s line in the plot using different colors and a legend?

plt.figure(figsize=(6,4),dpi=100)

plt.plot(df['Year'], df['Australia'], label='Australia')
plt.plot(df['Year'], df['Canada'], label='Canada')
plt.plot(df['Year'], df['France'], label='France')

plt.title('Inflation Rate Comparison (1990-1994)')
plt.xlabel('Year')
plt.ylabel('Inflation Rate (%)')

plt.legend()
plt.grid()
plt.show()
```

```python
# How would you add titles, labels to both axes, and a grid to the previous plot?

# Adding titles, labels, and a grid
plt.figure(figsize=(6, 4))

plt.plot(df['Year'], df['Australia'], marker='o', label='Australia')
plt.plot(df['Year'], df['Canada'], marker='s', label='Canada')
plt.plot(df['Year'], df['France'], marker='^', label='France')
plt.title('Inflation Rate Comparison with Labels (1990-1994)')

plt.xlabel('Year')
plt.ylabel('Inflation Rate (%)')

plt.legend(loc=2)
plt.grid(True)
plt.show()
```

```python
# Create a 2x2 grid of subplots where each subplot shows the inflation rate for one of these countries:
# USA, Japan, Germany, and Italy. How do you adjust the space between subplots?

plt.figure(figsize=(8,6),dpi=100)

plt.subplot(2,2,1)
plt.plot(df["Year"],df["USA"],'r.-',label="aus")
plt.legend()

plt.subplot(2,2,2)
plt.plot(df["Year"],df["Japan"],'b.-',label="can")
plt.legend()

plt.subplot(2,2,3)
plt.plot(df["Year"],df["Germany"],'g.-',label="fran")
plt.legend()

plt.subplot(2,2,4)
plt.plot(df["Year"],df["Italy"],'g.-',label="italy")
plt.legend()

# plt.tight_layout()
plt.legend()

plt.show()
```

```python
# On the line plot for the USA, annotate the highest inflation point on the graph. 
# How would you add an arrow pointing to this point with a text annotation?

plt.figure(figsize=(6,4),dpi=100)
plt.plot(df["Year"],df["USA"],'r.-',label="aus")

value=df["USA"].max()
year=df.loc[df["USA"].idxmax(), "Year"]

plt.annotate("Max Point",xy=(year,value),xytext=(year-6,value-0.5),
            arrowprops=dict(facecolor="black",arrowstyle='->'))

plt.legend()
plt.show()
```

```python
# Increase the size of the plot window and apply a specific Matplotlib style (e.g., ggplot). How do you achieve this?

plt.style.use("ggplot")

plt.figure(figsize=(6,4),dpi=100)
plt.plot(df["Year"],df["USA"],'r.-',label="aus")

value=df["USA"].max()
year=df.loc[df["USA"].idxmax(), "Year"]

plt.annotate("Max Point", 
             xy=(year, value), 
             xytext=(year - 6, value - 0.5),  
             arrowprops=dict(facecolor="black", edgecolor="black", linewidth=1.5, arrowstyle='->'))  

a=list(np.linspace(1990,2008,19))
plt.xticks(a[::2])

plt.legend()
plt.show()

# plt.style.available
```
